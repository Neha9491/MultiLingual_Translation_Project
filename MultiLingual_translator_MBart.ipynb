```python

from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
import gradio as gr

# Load mBART-50 multilingual model
model_name = "facebook/mbart-large-50-many-to-many-mmt"
tokenizer = MBart50TokenizerFast.from_pretrained(model_name)
model = MBartForConditionalGeneration.from_pretrained(model_name)

# Supported languages and language codes
language_dict = {
    "English": "en_XX",
    "Hindi": "hi_IN",
    "Kannada": "kn_IN",
    "Tamil": "ta_IN",
    "Telugu": "te_IN",
    "French": "fr_XX",
    "Spanish": "es_XX",
    "German": "de_DE",
    "Marathi": "mr_IN",
    "Gujarati": "gu_IN"
}
```

```python

# Translation function
def translate(text, source_lang, target_lang):
    tokenizer.src_lang = language_dict[source_lang]
    encoded = tokenizer(text, return_tensors="pt")
    generated_tokens = model.generate(
        **encoded,
        forced_bos_token_id=tokenizer.lang_code_to_id[language_dict[target_lang]],
        max_length=256,
        num_beams=4
    )
    return tokenizer.decode(generated_tokens[0], skip_special_tokens=True)
```

```python

# Gradio UI
interface = gr.Interface(
    fn=translate,
    inputs=[
        gr.Textbox(label="Enter Text"),
        gr.Dropdown(choices=list(language_dict.keys()), label="Source Language", value="English"),
        gr.Dropdown(choices=list(language_dict.keys()), label="Target Language", value="Kannada"),
    ],
    outputs=gr.Textbox(label="Translated Text"),
    title="🌍 Multilingual Translator (mBART-50)",
    description="Translate between 50+ languages using mBART-50. Powered by Hugging Face Transformers."
)

interface.launch()
```
